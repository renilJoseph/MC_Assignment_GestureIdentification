#MC Data Cleaning -- Data Preparation



## Imports

```
import pandas as pd
import numpy as np
```

```
! wget -q 'https://raw.githubusercontent.com/renilJoseph/MC_Assignment_GestureIdentification/master/data.csv'

df = pd.read_csv('data.csv')
df.head()
```

```
#Add vid_id

df['vid_id'] = 0
vid_id = -1
for row_idx, row in df.iterrows():
    if row['Frames#'] == 0:
        vid_id += 1
    df.loc[row_idx, 'vid_id'] = vid_id

df.to_csv('data_with_vid_id.csv', index=False)
```

## Prepare aggregate data

```
skip_cols = ['Frames#', 'type']

agg_data = df.drop(skip_cols,axis=1).groupby('vid_id').agg([np.mean, np.std])
agg_data.columns = ["_".join(x) for x in agg_data.columns.ravel()]
agg_data = agg_data.reset_index()
```

```
labels = df[['vid_id', 'type']].drop_duplicates()
```

```
agg_data = pd.merge(agg_data, labels, on=['vid_id'])
```
