#MC Data Cleaning -- Data Preparation



## Imports

```
import pandas as pd
import numpy as np
```

```
! wget -qN 'https://raw.githubusercontent.com/renilJoseph/MC_Assignment_GestureIdentification/master/good_data.csv'

df = pd.read_csv('good_data.csv')
df.head()
```

```
#Add vid_id

df['vid_id'] = 0
vid_id = -1
for row_idx, row in df.iterrows():
    if row['Frames#'] == 0:
        vid_id += 1
    df.loc[row_idx, 'vid_id'] = vid_id

df.to_csv('good_data_with_vid_id.csv', index=False)
```

## Prepare aggregate data

```
#merged_df = df.append(df_good, ignore_index=True, sort=True, verify_integrity=True)
labels = df[['vid_id', 'type']].drop_duplicates()
```

```
skip_cols = ['Frames#', 'type']
agg_data = df.drop(skip_cols,axis=1).groupby('vid_id').agg([np.mean, np.std, np.min, np.max, pd.DataFrame.kurt, pd.DataFrame.skew])
agg_data.columns = ["_".join(x) for x in agg_data.columns.ravel()]
agg_data = agg_data.reset_index()
```

```
agg_data = pd.merge(agg_data, labels, on=['vid_id'])
print(agg_data.shape)
agg_data.to_csv('agg_data.csv', index=False)
```
